```r
tracks_atac = c('jk.epipcg.atac.CRJK_0389_atac_wt_to_wt_eb_d3','jk.epipcg.atac.CRJK_0373_atac_wt_to_wt_epi_eb_d5',
'jk.epipcg.atac.CRJK_0377_atac_wt_to_wt_meso_eb_d5','jk.epipcg.atac.CRJK_0376_atac_wt_to_wt_endo_eb_d5')


tracks_f = tracks_atac

atac_meta = mod$track_atac_ribo
atac_meta = atac_meta[atac_meta$tr_atac %in% tracks_f, ,drop=F]
atac_meta = atac_meta[match(tracks_f, atac_meta$tr_atac), ]
ribo = atac_meta$ribo_c
tr_dsamp_r = min(ribo)/ribo
names(tr_dsamp_r) = atac_meta$short_name


#print(tr_dsamp_r)
for (tr in atac_meta$tr_atac) {
        gvtrack.create(atac_meta[atac_meta$tr_atac == tr, ]$short_name,
            tr, "sum")
        #gvtrack.iterator(atac_meta[atac_meta$tr_atac == tr, ]$short_name,
        #    sshift = -140, eshift = 140)
    }



meta_ucsc = data.frame(track = atac_meta$tr_atac,cell_type_id=1:nrow(atac_meta), 
           cell_type = atac_meta$short_name,
           color = c('#635547','#635547','#C594BF','#EF5A9D'), 
           shorthand=atac_meta$short_name)

###downsampled smoothed celltype tracks  




df <- meta_ucsc[ meta_ucsc$track %in% atac_meta$tr_atac,]
plyr::a_ply(df, 1, function(x){
       
        r_ds = tr_dsamp_r[ x$shorthand]
        print(r_ds)
        data <- gextract(x$shorthand, gintervals.all(),iterator = 300, colnames = "score")
        data$score[ is.na(data$score)] = 0
        to_ds = matrix(data$score, ncol = 1)
        post_ds = downsample_matrix(to_ds, target_n = floor(sum(to_ds) *
                  r_ds), seed = 19)
        data$score = post_ds
        data$score = log2(data$score+1)
        print(quantile(data$score,c(.2,.5,.98,.99,.995)))
        fwrite_ucsc(data, file = glue::glue("~/raid_dir/Proj/polycomb/seq2epi_paper_norm/ucsc_tracks//{x$shorthand}.ucsc"),
                   name = x$shorthand, description = x$shorthand, type="wig", span = 20, color = x$color)
    })
```

```r
cg_trace = readRDS('/net/mraid20/ifs/wisdom/tanay_lab/tgdata/users/evghenic/Proj/polycomb/seq2epi_paper_norm/data/cg_trace_mm10.rds')
f_norp = cg_trace$d_ltr != 0 & cg_trace$d_line != 0 & cg_trace$start > 3e+06
cg_trace_f = cg_trace[f_norp,c('chrom','start','end')]

dim(gintervals.canonic(cg_trace_f))
dim((cg_trace_f))
cg_trace_f = gintervals.canonic(cg_trace_f)

cg_trace_f$score = 1
cg_trace_f$intervalID = 1:nrow(cg_trace_f)
tr =  'f_repeat'   
color = 'purple'
df = cg_trace_f


# df has columns: chrom, start, end, score, intervalID

# Convert to UCSC BED (0-based start)
bed <- data.frame(
  chrom = as.character(df$chrom),
  chromStart = as.integer(df$start) - 1,   # UCSC BED: 0-based start
  chromEnd = as.integer(df$end),
  name = paste0("interval_", df$intervalID),
  score = pmax(0L, pmin(1000L, as.integer(round(df$score * 1000)))) # UCSC score 0-1000
)

# If your df$score is already fine as 0-1000, use this instead:
# bed$score <- pmax(0L, pmin(1000L, as.integer(df$score)))

out_bed <- "./ucsc_tracks/f_norp_ucsc_track.bed"

# Write with a UCSC track line at the top
track_line <- 'track name="f_norp" description="f_norp" visibility=2 itemRgb="On"'

con <- file(out_bed, open = "wt")
writeLines(track_line, con)
write.table(bed, con, sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)
close(con)

message("Wrote: ", normalizePath(out_bed))
```
